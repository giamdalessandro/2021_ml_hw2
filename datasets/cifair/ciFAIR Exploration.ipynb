## Visualization

```python
data = ciFAIR10(data_root, 'train')
len(data)
```

```python
fig, axes = plt.subplots(data.num_classes, 10, figsize=(13, 1.5*data.num_classes))
for lbl, (classname, row) in enumerate(zip(data.classes, axes)):
    row[0].set_ylabel(classname)
    for img, ax in zip(data.data[data.targets == lbl], row):
        ax.imshow(img)
        ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
fig.tight_layout()
plt.show()
```
